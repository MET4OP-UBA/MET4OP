# VISUALIZACIÓN DE DATOS

+ Herramienta de representación gráfica de datos, como forma accesible de comprenderlos 

## MATPLOTLIB VS SEABORN: ¿por qué y para qué?

+ Desventajas de Matplotlib
    + solo permite graficos simple
    + sintaxis complicada y extensa
+ Ventajas de Seaborn
    + Es una libreria construida a partir de Matplotlib
    + sintaxis intuituva y simple
    + permite visualizar graficos mas complejos 
    + Used for plotting statistical graphs, identifying trends, relationships & outliers
    
## Indice 
+ Datos numericos:
    + relplot()
    + lineplot()
    + scatterplot()
+ Datos categóricos:
    +  catplot()
    + boxplot()
+ Visualizacion de la distribución:
    + histplot()
    +  displot()
    +  kdeplot()
+ Regresión lineal:
   + lmplot()
   + regplot()
+ Grillas multiples
   + pairplots()
   + joinplots()
+ Matrix/color:
   + heatplot()


```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline #para que los graficos aparezcan sin espesificar .show()
```

```python
fmri = sns.load_dataset('fmri') #ejemplo de dataset automatico de las librerias
fmri.head()
```

```python
futbol = pd.read_csv("../54929/MET4OP/dataset/extras/datos_2015.csv")
futbol.head()
```

```python
#RELPLOT
#Los diagramas de dispersión visualizan cada tupla de valores x e y como un solo punto y el diagrama formará una nube de puntos
sns.set(style = 'darkgrid')

sns.relplot(x = 'Pts_2014', y = 'club', data= futbol)
```

```python
dir(sns.FacetGrid) #Matplotlib admite la creación de figuras con múltiples ejes y permite tener subtramas en una figura
```

```python
#funcion HUE: perimite incorporar una variable que agrupa los datos ya importados
    #Los diferencia por color
    # Puede ser numerica o categórica

sns.relplot(x = 'Pos_Fecha_30', y = 'club', data= futbol, hue= 'Pos_Fecha_21' , style = 'ascenso')
```

```python
#Podemos separar los datos en columnas o filas 
ax = sns.relplot(x='valor' , y='inversion_relativa',   row = 'ascenso', data= futbol)
```

```python
#SCATTERPLOT
#estudiar la asociación de dos variables de intervalo, resumen útil de un grupo de datos bivariados
ax = sns.scatterplot(x='valor' , y='club' ,  hue = 'inversion_relativa', data= futbol , size = 'inversion_abs',
                    sizes=(20, 100))
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line' , data= fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line' , data= fmri , ci = False)
```

```python
sns.relplot(x = 'valor', y = 'inversion_relativa', kind = 'line' , hue = 'ascenso' , data= futbol)
```

```python
#LINEPLOT, en lugar de especificar el kind = 'line' vamos a introducir la funcion sns.lineplot()
ax = sns.lineplot(x='valor', y='inversion_relativa', data = futbol)
```

```python
ax = sns.lineplot(x='valor' , y='inversion_relativa', data = futbol , hue= 'ascenso' , style= 'ascenso')
```

```python
ax = sns.lineplot(x='valor' , y='inversion_relativa', data = futbol , hue= 'ascenso' , style= 'ascenso' , markers=True)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#CATPLOT, muestra la relacion entre una variable numerica y una o mas variables categoricas 
sns.catplot(x = 'day' , y = 'total_bill' , data = tips)
```

## Visualización de la distribucion de datos

```python
x = futbol['inversion_abs'].values
x
```

```python
#histogramas para ver la distribucion de frecuencias
x = futbol['inversion_abs'].values
sns.distplot(x, color = 'blue' , rug= True);
```

```python
#como sacar la media 
x = futbol['inversion_abs'].values
sns.histplot(x, color = 'blue', kde=True);


mean = futbol['inversion_abs'].mean()
plt.axvline(mean, 0,1, color = 'red') #especifica que muestre la media como una linea vertical sobre el histplot
```

```python
#KDE: kernel density estimation. Es analoga a los histplot y grafica una curva de densidad continua.
sns.kdeplot(x , shade= True)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=0)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="stack")
```

```python
#cuando el hue es numerico
sns.kdeplot(data=tips, x="total_bill", hue="size")
```

```python
sns.kdeplot(
   data=tips, x="total_bill", hue="size",
   fill=True, common_norm=False , palette="crest",
   alpha= 0.5, linewidth=0,
  ) 
```

## Regresiones lineales 

+ Regplot () se puede considerar como un subconjunto de las características de lmplot ()
+ La diferencia entre regplot () y lmplot () es que cuando desea agregar una tercera o más variables, solo puede usar la última; porque lmplot () combina regplot () y FacetGrid



```python
ax = sns.lmplot(x="valor", y="inversion_relativa", data=futbol, height=9)
```

```python
ax = sns.lmplot(x="valor", y="inversion_relativa", data=futbol, col = 'ascenso') #row
```

```python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", truncate=False , 
                markers=["o", "x"], data=tips)
```

```python
sns.regplot(x = 'total_bill', y = 'tip', data= tips,)
# hue= 'sex'
```

## Grillas multiples

```python
#Configurar una grilla con matplotlib y especificar que datos van en cada uno
fig, axes = plt.subplots(2, 2, figsize=(12, 7))

a = futbol['valor'].values
b = futbol['inversion_relativa'].values
c = futbol['inversion_abs'].values


# plot a
sns.histplot(a, color = 'blue', ax=axes[0,0])

# plot b
sns.histplot(b, color = 'blue', ax=axes[0,1])

# plot c
sns.histplot(c, color = 'blue', ax=axes[1,0])

# plot d
sns.boxplot(x="Pts_2014", y="Pos_Fecha_30", data=futbol, color = 'green', ax=axes[1,1])
```

```python
futbol = pd.read_csv("../54929/MET4OP/dataset/extras/datos_2015.csv")
futbol.head()
```

```python
#esta función creará una cuadrícula de ejes: y en una sola fila y los ejes x en una sola columna. 
#diagonales: se dibuja una gráfica de distribución univariante para mostrar la distribución marginal de los datos en cada columna
g = sns.pairplot(futbol[['Pts_2014','inversion_abs','inversion_relativa','valor','Pos_Fecha_30','ascenso']], plot_kws={'color':'red'})
```

```python
g = sns.pairplot(futbol[['valor','inversion_abs','inversion_relativa','Pts_2014','Pos_Fecha_30', 'ascenso']], hue= 'ascenso')
```

```python
g = sns.pairplot(futbol[['valor','inversion_abs','inversion_relativa','Pts_2014','Pos_Fecha_30', 'ascenso']], 
                 hue= 'ascenso', kind= 'reg')
```

```python
#dos variables con gráficas bivariadas y univariadas
g = sns.jointplot("inversion_abs", "inversion_relativa", data=futbol, kind="reg", color = 'green', size = 10)
```

```python
g = sns.jointplot("inversion_abs", "valor", data=futbol, kind="kde", color = 'green', fill= True, size = 10)
```

## Matrix/colores

```python
#Trazar datos rectangulares como una matriz codificada por colores
pc = futbol[['inversion_abs','inversion_relativa','valor', 'ascenso']].corr(method ='pearson')

cols = ['inversion_abs CC','inversion_relativa CC','valor CC']

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 50})
```
