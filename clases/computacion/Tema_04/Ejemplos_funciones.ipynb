# Funciones en Python

## **Funciones a medida**

```python
def sumcount(n):
    '''
    Devuelve la suma de los primeros n enteros
    '''
    total = 0
    while n > 0:
        total += n
        n -= 1
    return total
```

```python
sumcount(10)
```

## **Funciones de la biblioteca**

```python
import math

def raiz_cuadrada(*args):
    '''
    Devuelve la raíz cuadrada de todos los argumentos que ingrese el usuario
    '''
    for i in args:
        print("La raiz cuadrada de", i,  "es:" , math.sqrt(i))
```

```python
raiz_cuadrada(144, 169)
```

## **Función de ordenamiento: bubble sort**

```python
def bubbleSort(lista): 
    n = len(lista) 
  
     # Recorre todos los elementos de la matriz
    for i in range(n): 
        print("iteración", i, ":", lista)
         # Los últimos elementos i ya están en su lugar
        for j in range(0, n-i-1):   
            # atraviesa la matriz de 0 a n-i-1
             # Intercambiar si el elemento encontrado es mayor
             # que el siguiente elemento 
            if lista[j] > lista[j+1] : 
                lista[j], lista[j+1] = lista[j+1], lista[j]
```

```python
opinion = [10,1,0,7,3,5,2,7,5,8,9,2,4,6,9,1,2,8]
bubbleSort(opinion)
```
