### Criterio de Información de Akaike (AIC) en una Regresión Lineal

El **Criterio de Información de Akaike (AIC)** es una medida para evaluar modelos estadísticos, comúnmente utilizada en regresiones lineales.  
El AIC estima la calidad de un modelo en función de su ajuste a los datos y la complejidad del modelo, con el objetivo de identificar el modelo que mejor generaliza.

\[
AIC = 2k - 2 \ln(\hat{L})
\]

donde:
- \( k \) es el número de parámetros en el modelo,
- \( \hat{L} \) es el valor máximo de la función de verosimilitud del modelo.

**Interpretación**: Al comparar modelos que se ajustan a los mismos datos, el modelo con el **AIC más bajo** es el preferido,  
ya que indica un mejor balance entre el ajuste y la simplicidad del modelo.

---

### Criterio de Información Bayesiano (BIC) en una Regresión Lineal

El **Criterio de Información Bayesiano (BIC)** es una medida similar al AIC, también utilizada para seleccionar modelos en función de su ajuste  
y complejidad, pero con una penalización más fuerte para modelos con mayor número de parámetros.

\[
BIC = k \ln(n) - 2 \ln(\hat{L})
\]

donde:
- \( n \) es el número de observaciones en el conjunto de datos,
- \( k \) es el número de parámetros en el modelo,
- \( \hat{L} \) es el valor máximo de la función de verosimilitud del modelo.

**Interpretación**: Al comparar modelos, el **BIC más bajo** es preferible, ya que indica un modelo que equilibra el ajuste y la simplicidad,  
aunque penaliza más fuertemente los modelos complejos en comparación con el AIC, favoreciendo así modelos más parsimoniosos.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define el modelo
model = LinearRegression()

# Define la matriz de predictores (X) y la variable respuesta (y)
X = data2010.drop(columns=['Life expectancy '])  # Asegúrate de que 'Life expectancy' es tu variable objetivo
y = data2010['Life expectancy ']

# Configura Recursive Feature Elimination (RFE)
selector = RFE(estimator=model, n_features_to_select=5, step=1)  # Puedes ajustar n_features_to_select
selector = selector.fit(X, y)

# Muestra las variables seleccionadas
selected_features = X.columns[selector.support_]
print("Selected features:", selected_features)

# Crea un nuevo modelo usando solo las variables seleccionadas
X_selected = X[selected_features]
model.fit(X_selected, y)

# Evalúa el modelo
y_pred = model.predict(X_selected)
mse = mean_squared_error(y, y_pred)
```

```python

# Crea un nuevo modelo usando solo las variables seleccionadas
X_selected = X[selected_features]
model.fit(X_selected, y)

# Evalúa el modelo
y_pred = model.predict(X_selected)
mse = mean_squared_error(y, y_pred)
print(mse)
```

El **Variance Inflarion Factor** (VIF) es una medida que evalúa la colinealidad entre variables independientes en un modelo de regresión.  
Un VIF alto indica que una variable está altamente correlacionada con otras variables independientes, lo cual puede inflar  
la varianza de los coeficientes y hacer que el modelo sea inestable o que sus coeficientes sean difíciles de interpretar.

### Interpretación básica:
- **VIF = 1**: No hay colinealidad.
- **VIF entre 1 y 5**: Colinealidad moderada; generalmente aceptable.
- **VIF > 5 o 10**: Colinealidad alta; podría ser problemático y sugiere considerar la eliminación o combinación de variables. 

Un VIF elevado en una variable puede implicar que los cambios en los coeficientes se deben a redundancias con otras variables,  
afectando la confiabilidad del modelo.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Excluir la constante antes de calcular el VIF
X_no_const = X.drop(columns=['const'])

# Calcular el VIF para cada columna en X_no_const
vif = [variance_inflation_factor(X_no_const.values, i) for i in range(X_no_const.shape[1])]

# Crear un DataFrame con los VIFs y los nombres de las columnas sin la constante
vif_model4 = pd.DataFrame({"VIF": vif}, index=X_no_const.columns)

vif_model4
```

### Apalancamiento o Leverage points

### Puntos de Influencia en una Regresión Lineal

Los **puntos de influencia** en una regresión lineal son observaciones que tienen un impacto significativo en el ajuste del modelo. Estos puntos pueden alterar la pendiente, el intercepto o la estructura general de la regresión, y suelen detectarse mediante la métrica de **leverage** y otras herramientas de diagnóstico.

Un punto con **leverage alto** tiene una distancia considerable respecto al centro de los datos en el espacio de las variables predictoras, lo cual le otorga un peso elevado en la determinación de los coeficientes del modelo.

### Determinación de Leverage Alto

El **leverage** para cada observación \(i\) se denota como \( h_{ii} \), y un valor de leverage se considera alto si cumple la siguiente condición:

$$
h_{ii} > \frac{2p}{n}
$$

donde:
- \( p \) es el número de parámetros en el modelo, incluyendo el intercepto,
- \( n \) es el número total de observaciones en el conjunto de datos.

**Interpretación**: Si el leverage de una observación excede el valor de \( \frac{2p}{n} \), esta observación se considera un punto de leverage alto y puede tener una influencia considerable en el ajuste del modelo. Es recomendable analizar estos puntos de manera más profunda, ya que podrían ser outliers o representar datos atípicos en el espacio de los predictores.

```python
fig = sm.graphics.influence_plot(model3, criterion="cooks")
fig.tight_layout(pad=1.0)
fig.show()
```

```python
# Calcula las influencias del modelo
influence = model3.get_influence()

# Obtiene el valor de Cook's distance
cooks_d = influence.cooks_distance[0]

# Establece un umbral para identificar puntos de influencia (por ejemplo, 4/n)
threshold = 4 / len(data2010)

# Identifica las observaciones que son puntos de influencia
influential_points = X[cooks_d > threshold]

# Imprime las observaciones que son puntos de influencia
print("Observaciones que son puntos de influencia:")
print(influential_points)
```

```python
plt.scatter(X["Income composition of resources"],cooks_d)
plt.xlabel("X")
plt.ylabel("Cooks Distance")
plt.show()
```

### D de Cook

La **D de Cook** es una medida en regresión lineal que identifica observaciones con un impacto desproporcionado en el ajuste del modelo. Cuanto mayor sea el valor de D de Cook para una observación, mayor es su influencia sobre los coeficientes del modelo.

#### Valor Crítico
Una observación se considera influyente si su **D de Cook es mayor a \( 4/n \)**, donde \( n \) es el número total de observaciones.
